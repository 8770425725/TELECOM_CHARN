## **PROBLEM STATMENT**
Telecom Churn  is a problem for telecom companies because it is expensive to acquire a new customer and companies want to retain their existing customers. Most telecom companies suffer from voluntary churn.

Customer churn is when a company’s customers stop doing business with that company. Businesses are very keep on measuring churn because keeping an existing customer is far less expensive than acquiring a new customer. New business involves working leads through a sales funnel, using marketing and sales budgets to gain additional customers. Existing customers will often have a higher volume of service consumption and can generate additional customer referrals.

## **BUSINESS OBJECTIVE**
The business objective is to predict the churn in the last month using the data from the first three months. To do this task well, understanding the typical customer behaviour during churn will be helpful.
To reduce customer churn, telecom companies need to predict which customers are at high risk of churn.

## **Let's Analyzing State column**

```python
#lets print the unique value of column
telecom_df['State'].nunique()
```

```python
#comparision churn with state by using countplot
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
ax=sns.countplot(x="State",hue='Churn',data=telecom_df)
plt.show()
```

```python
s1=telecom_df["State"].unique()
s2=telecom_df.groupby(['State'])["Churn"].mean()
plt.rcParams['figure.figsize']=(15,7)
plt.plot(s1,s2,color="r",marker='o',linewidth=2,markersize=12)
plt.xlabel('state',fontsize=15)
plt.ylabel('churn rate',fontsize=15)
plt.title('State churn rate',fontsize=20)
plt.show()
```

```python
#calculate state vs churn persentage
State_data=pd.crosstab(telecom_df['State'],telecom_df['Churn'])
State_data['persentage_churn']=State_data.apply(lambda x:x[1]*100/(x[0]+x[1]),axis=1)
State_data
```

```python
#lets see the lowest to 10 churn state by ascending order
telecom_df.groupby(['State'])['Churn'].mean().sort_values(ascending=False).head(10)
```

There is 51 unique state present who have diffrent churn rate.

From the above analysis CA,NJ,TX,MD,SC,MI are the ones who have a higher churn rate of more then 51.

The reason for this churn rate from a particular state may be due to the low coverage of the cellular network.
